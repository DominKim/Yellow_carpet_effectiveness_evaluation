```python
import pandas as pd
import numpy as np
```

```python
# 2019_03

df2019_03 = pd.read_csv("./유동인구/FLT_SEOUL_03MONTH.csv", encoding = "utf-8")

gu_lst = list(df2019_03["군구"].unique())
days = list(df2019_03["일자"].unique())

for idx in range(len(df2019_03)):
    df2019_03[df2019_03["군구"]]


use_dict_2019_03 = {"일자" :[], "군구":[], "유동인구수":[]}


for day in days:
    for gu in gu_lst:
        people = sum(df2019_03[(df2019_03["일자"] == day) & (df2019_03["군구"] == gu)]["유동인구수"])
        use_dict_2019_03["일자"].append(day); use_dict_2019_03["군구"].append(gu); use_dict_2019_03["유동인구수"].append(people)
        
        
people_2019_03 = pd.DataFrame(use_dict_2019_03)
```
