```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv("./qgis_data/2019_전처리.csv", encoding = "euc-kr")
df["옐로카펫_"] = df["옐로카펫_"].str.replace("-", "")
```

```python
weather = pd.read_csv("./기상정보_20100101-20191231.csv", encoding = "euc-kr")
```

```python
weather["날짜"] = weather["날짜"].str.replace("-", "")
```

```python
weather.head(1)
```

```python
df = pd.merge(df, weather, left_on = "발생년월일", right_on="날짜", how = "inner")
```

```python
df["옐로카펫_설치유무"] = 0
```

```python
lst_idx = df["옐로카펫_"].dropna().index

for idx in lst_idx:
    if df.loc[idx, "옐로카펫_"] != "0":
        if df.loc[idx, "발생년월일"] > df.loc[idx, "옐로카펫_"]:
            df.loc[idx, "옐로카펫_설치유무"] = 1
```

```python
df.to_csv("./qgis_data/2019_전처리_기상정보.csv", encoding = "euc-kr", index = False)
```

```python
df = pd.read_csv("./qgis_data/2018.csv", encoding = "euc-kr")
df["옐로카펫_"] = df["옐로카펫_"].str.replace("-", "")
df["발생년월일"] = df["발생년월일"].astype(object)
df["평균기온(℃)"] = 0; df["최저기온(℃)"] = 0; df["최고기온(℃)"] = 0; df["강수량(mm)"] = 0

weather = pd.read_csv("./기상정보_20100101-20191231.csv", encoding = "euc-kr")

weather["날짜"] = weather["날짜"].str.replace("-", "")

for idx in range(len(df)):
    a = weather[weather["날짜"] == str(df.loc[idx, "발생년월일"])]

    df.iloc[idx, 39] = a.iloc[0, 2]
    df.iloc[idx, 40] = a.iloc[0, 3]
    df.iloc[idx, 41] = a.iloc[0, 4]
    df.iloc[idx, 42] = a.iloc[0, 5]

df["옐로카펫_설치유무"] = 0


lst_idx = df["옐로카펫_"].dropna().index

for idx in lst_idx:
    if df.loc[idx, "옐로카펫_"] != "0":
        if str(df.loc[idx, "발생년월일"]) > str(df.loc[idx, "옐로카펫_"]):
            df.loc[idx, "옐로카펫_설치유무"] = 1
```

```python
df.to_csv("./qgis_data/2018_전처리_기상정보.csv", encoding = "euc-kr")
```
